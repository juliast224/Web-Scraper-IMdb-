**Question 2.9** (5 points) Creat a new column ROI that measures the return of investment using (box revenue-budget)/budget, and compare the average ROI between movies that passed (rating=3) and failed the test (rating=0), are their ROI different? Explain.

````python
budget_df = bechdel_imdb_top

# drop the columns that do not start with $
budget_df = budget_df[budget_df['budget'].str.startswith('$')]
budget_df = budget_df[budget_df['gross_box_office_rev'].str.startswith('$')]

# remove the $ sign of budget and gross box office revenue columns and change to integer        
budget_df['budget_without_currency'] = budget_df['budget'].str[1:]
budget_df['budget_without_currency'] = budget_df['budget_without_currency'].str.replace(',', '').astype(int)

budget_df['gross_box_office_rev_wc'] = budget_df['gross_box_office_rev'].str[1:]
budget_df['gross_box_office_rev_wc'] = budget_df['gross_box_office_rev_wc'].str.replace(',', '').astype(int)

# create new column for ROI and round the values
budget_df['ROI'] = round((budget_df['gross_box_office_rev_wc']/budget_df['budget_without_currency']), 2)
passed_roi = round((budget_df[budget_df['rating_y'] == 3]['ROI'].mean()), 2)
failed_roi = round((budget_df[budget_df['rating_y'] == 0]['ROI'].mean()), 2)
print('The average ROI of the passed movies is:',passed_roi,'\n','The average ROI of the failed movies is: ', failed_roi)
````
